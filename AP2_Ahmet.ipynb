```python
import googletrans
from googletrans import *
```

```python
translator = googletrans.Translator()
```

```python
required_df["食品価格動向調査（野菜）の結果（価格、単位：円/kg）"] = required_df["食品価格動向調査（野菜）の結果（価格、単位：円/kg）"].astype(str)
required_df["食品価格動向調査（野菜）の結果（価格、単位：円/kg）"] = required_df["食品価格動向調査（野菜）の結果（価格、単位：円/kg）"].apply(translator.translate, src='ja', dest='en').apply(getattr, args =('text',) )
required_df
```
